```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("amazon.csv")
data.head(10)
```

```python
data.info()
```

```python
data = data[["product_id","product_name", "category","discounted_price","actual_price","discount_percentage","rating","rating_count","user_name"]]
data.sample()
```

```python
data["rating"].unique() # displaying unique values for rating
```

```python
# Removing unwanted string value from a key value
data = data[data["rating"] != '|']
```

```python
# Unwanted string removed
data["rating"].unique()
```

```python
# Coverting object to float
data["rating"] = data["rating"].astype(float)
```

```python
# Removing the unwanted symbol from actual_price value in the column
data["actual_price"] = data["actual_price"].str.replace('₹', '').str.replace(',', '')
```

```python
# Actual price dtype to float
data["actual_price"] = pd.to_numeric(data["actual_price"])
```

```python
# Removing the unwanted symbol ',' from rating_count values in the column
data["rating_count"] = data["rating_count"].str.replace(',', '')
```

```python
# Checking and Dropping null values from rating_count
data["rating_count"].isnull().sum()
data.dropna(inplace=True)
```

```python
# Coverting rating_count object to int
data["rating_count"] = data["rating_count"].astype(float)
```

```python
# Removing the unwanted symbol '%' from discount_percentage values in the column
data["discount_percentage"] = data["discount_percentage"].str.replace('%', '')
```

```python
# Coverting discount_percentage object to int
data["discount_percentage"] = data["discount_percentage"].astype(int)
```

```python
# Removing the unwanted symbol '₹', ',' from discount_percentage values in the column
data["discounted_price"] = data["discounted_price"].str.replace('₹', '').str.replace(',', '')
```

```python
# Coverting discount_percentage object to int
data["discounted_price"] = data["discounted_price"].astype(object).astype(float)
```

```python
# Removing duplicate values from the dataset
print(data.duplicated().sum())
data.loc[data.duplicated()].head() # Displaying where duplicate exits
```

```python
data.loc[data.duplicated(subset= ["product_name"])]
```

```python
print(data.duplicated(subset= ["product_name"]).sum())
data.query("product_name == 'Firestick Remote'") # Getting duplicated roles by product_name
data.query("product_name == ''")
```

```python
# Removing duplicate values
data = data.loc[~data.duplicated(subset= ["product_name", "discounted_price", "actual_price","rating"])].reset_index(drop=True)
```

```python
# Verifying removed duplicate values
data.duplicated().sum()
data.isnull().sum()
```

```python
# Splitting category into two divisions 
data['main_division'] = data['category'].astype(str).str.split('|').str[0]
data['sub_division'] = data['category'].astype(str).str.split('|').str[-1]
```

```python
# Renaming the columns
data = data.rename(columns={'product_id': 'Product_Id','product_name': 'Product_Name', 'category': 'Category', 
                          'discounted_price': 'Discounted_Price','actual_price': 'Actual_Price',
                         'discount_percentage': 'Discount_Percentage', 'rating': 'Rating',
                          'rating_count': 'Rating_Count', 'user_id' : 'User_ID', 
                          'user_name':'User_Name', 'main_division' : 'Main_Division',
                          'sub_division' : 'Sub_Division'}, inplace = False)
```

```python
# Reformating the title descripion for groups in the Main_Division column. 
data['Main_Division'] = data['Main_Division'].str.replace('&', ' & ') 
data['Main_Division'] = data['Main_Division'].str.replace('Instruments', ' Instruments ') 
data['Main_Division'] = data['Main_Division'].str.replace('Office', ' Office ') 
data['Main_Division'] = data['Main_Division'].str.replace('Products', ' Products ')
data['Main_Division'] = data['Main_Division'].str.replace('Improvement', ' Improvement ') 
```

# CORRELATION ANALYSIS

```python
# Visualisation of correlation between Actual Price and Discounted Price

correlation = data[["Actual_Price", "Discounted_Price"]].corr()
plt.imshow(correlation, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.title("Correlation between Actual Price and Discounted Price")
plt.show()
```

```python
# Shows the correlation of the dtypes that excludes objects
data_correlation = data.select_dtypes(exclude=['object', 'category']).dropna().corr() 
data_correlation.head()
```

```python
# Heatmap for the above correlation
sns.heatmap(data_correlation, annot=True)
```

```python
# Distribution of rating score
rating_score = []
for score in data['Rating']:
    if score < 2.0:
        rating_score.append('Poor')
    elif score < 3.0:
        rating_score.append('Below Average')
    elif score < 4.0:
        rating_score.append('Average')
    elif score < 5.0:
        rating_score.append('Above Average')
    elif score == 5.0:
        rating_score.append('Excellent')
```

```python
data['rating_score'] = rating_score
data['rating_score'] = data['rating_score'].astype('category')
```

```python
rating_score = data['rating_score'].value_counts().reindex(index = ['Below Average', 'Average', 'Above Average', 'Excellent']).rename_axis('rating_score').reset_index(name='counts')
rating_score
```

```python
# Distribution of rating score

ax = plt.subplots(figsize=(8, 6))[1]
ax.bar(x=rating_score['rating_score'], height=rating_score['counts'])
ax.set_title('Distribution of rating score')
ax.set_xlabel('Rating score')
ax.set_ylabel('Counts')

plt.show()
```
